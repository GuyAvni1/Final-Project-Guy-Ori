# **Null Filling**
In this part, we choose the relevant features for training and fill the data set by using HistGradientBoostingRegressor model.

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
```

```python
#Reading the data set from the drive
raw = pd.read_csv("/content/drive/MyDrive/Final_Project_Data/df_prepared_after_ps.csv")
data = raw
data.head() #Showing an example of the first 5 rows.
```

```python
# Taking only the relevant features from the complete dataset
new_data = data[['RID', 'VISCODE'  ,'FDG', 'AV45', 'ABETA', 'TAU', 'PTAU', 'CDRSB', 'ADAS11',
                 'ADAS13', 'MMSE', 'RAVLT_immediate', 'RAVLT_forgetting',
                 'RAVLT_perc_forgetting', 'FAQ', 'MOCA', 'RAVLT_learning',
                 'Ventricles', 'Hippocampus', 'WholeBrain', 'Entorhinal', 'Fusiform', 'MidTemp', 'ICV', 'DX']]
new_data.dropna(subset = ['DX'], inplace=True)
new_data.head()
```

```python
# Showing there are multiple nulls values in the dataset
new_data.dropna(subset = ['DX'], inplace=True)
labels = new_data[['RID','VISCODE','DX']]
print(new_data.isna().sum())
print(labels.shape)
new_data.shape
labels
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.impute import SimpleImputer

data = new_data
data.drop(columns=['DX','VISCODE','RID'], axis = 1,inplace = True)

# Iterate through each column with missing values
for column in data.columns:
    if data[column].isnull().any():  # Check if column has missing values
        # Create a copy of the data to work with
        impute_data = data.copy()

        # Split data into features and target (column with missing values)
        features = impute_data.drop(column, axis=1)
        target = impute_data[column]

        # Remove rows where the target has missing values
        non_missing_indices = target.dropna().index
        features = features.loc[non_missing_indices]
        target = target.loc[non_missing_indices]

        # Split data into train and test sets
        X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

        # Train a machine learning model (HistGradientBoostingRegressor)
        model = HistGradientBoostingRegressor()
        model.fit(X_train, y_train)

        # Use the trained model to predict missing values
        missing_indices = data[data[column].isnull()].index
        predicted_values = model.predict(data.loc[missing_indices].drop(column, axis=1))

        # Fill in the missing values with predicted values
        data.loc[missing_indices, column] = predicted_values
```

```python
# Checking for remaining nulls
data.isna().sum()
```
