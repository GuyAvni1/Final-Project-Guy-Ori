# **Data After Participant Selection**
Creating df_prepared file with the correct patients, known from df_ps

```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files

# Loading the df_prepared.csv and df_ps.csv files
df1 = pd.read_csv("/content/drive/MyDrive/Final_Project_Data/df_prepared.csv")
data = df1
df2 = pd.read_csv("/content/drive/MyDrive/Final_Project_Data/df_ps.csv")
array1 = df1['RID'].unique()
array2 = df2['RID'].unique()

# Find the diffrence between the patients in the files (by RID column)
diff_values = np.setdiff1d(array1, array2)

# Remove the unnecessaries rows (unnecessaries patiens)
drop_rows = data[data['RID'].isin(diff_values)].index
data.drop(drop_rows, inplace=True)

# Making sure the vlaues are sorted and creating the df_prepared_after_ps.csv
data.sort_values(by=['RID','Month_bl'], inplace=True)
data.to_csv('df_prepared_after_ps.csv', index=False)
files.download('df_prepared_after_ps.csv')

```
